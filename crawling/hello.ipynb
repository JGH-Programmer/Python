```python
import requests

url = "https://www.daum.net/"
response = requests.get(url)

# HTTP 응답 상태 코드 확인
print(response.status_code)  # 200 (성공)

# 응답 본문(HTML 소스) 확인
print(response.text)
```

```python
html_content = response.text
start = html_content.find('<title>')
end = html_content.find('</title>')

title_content = html_content[start + len('<title>'): end]
print(title_content)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# 브라우저 열기
driver = webdriver.Chrome()

# 웹 페이지 로드
url = '<http://example.com>'
driver.get(url)

# JavaScript 실행 대기
time.sleep(3)  # 페이지 로드 및 JavaScript 실행을 위한 대기 시간

# 요소 찾기
heading = driver.find_element(By.ID, 'main-title').text
print("Heading:", heading)

# 브라우저 닫기
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup

# chrome driver 설정
driver = webdriver.Chrome()
```

```python
# URL 열기
url = 'https://booking.naver.com/booking/12/bizes/181387/items/2870248'
driver.get(url)
# 페이지 로딩 대기 (필요 시)
driver.implicitly_wait(10)  # 최대 10초 대기
```

```python
# 날짜 버튼들 찾기
calendar_dates = driver.find_elements(By.CSS_SELECTOR, 'a.calendar-date')
len(calendar_dates)
```

```python
date_button = calendar_dates[0]
date_button.click()
```

```python
# 날짜별 예약 시간대 정보 저장할 리스트
date_time_info = [0]
```

```python
for date_button in calendar_dates:
    try:
        # 날짜 버튼 클릭
        date_button.click()
        
        # 페이지 로드 시간 대기 (필요 시 조정)
        time.sleep(2)  # 페이지 로드 시간 대기
        driver.implicitly_wait(2)
        html_code = driver.page_source
    except:
        print("오류발생")
```
