```python
import requests

url = "https://www.daum.net/"
response = requests.get(url)

# HTTP 응답 상태 코드 확인
print(response.status_code)  # 200 (성공)

# 응답 본문(HTML 소스) 확인
print(response.text)
```

```python
html_content = response.text
start = html_content.find('<title>')
end = html_content.find('</title>')

title_content = html_content[start + len('<title>'): end]
print(title_content)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup

# chrome driver 설정
driver = webdriver.Chrome()
```

```python
# URL 열기
url = 'https://booking.naver.com/booking/12/bizes/181387/items/2870248'
driver.get(url)
# 페이지 로딩 대기 (필요 시)
driver.implicitly_wait(10)  # 최대 10초 대기
```

```python
# 날짜 버튼들 찾기
calendar_dates = driver.find_elements(By.CSS_SELECTOR, 'a.calendar-date')
len(calendar_dates)
```

```python
date_button = calendar_dates[0]
date_button.click()
```

```python
# 날짜별 예약 시간대 정보 저장할 리스트
date_time_info = [0]
```

```python
import requests

url = "https://www.10000recipe.com/ranking/home_new.html"
response = requests.get(url)

# HTTP 응답 상태 코드 확인
print(response.status_code)  # 200 (성공)

# 응답 본문(HTML 소스) 확인
print(response.text)
```

```python
import requests
from bs4 import BeautifulSoup

# 웹 페이지의 URL
url = 'https://www.10000recipe.com/ranking/home_new.html'

# 해당 URL에 HTTP 요청 보내고 HTML 가져오기
response = requests.get(url)
html_content = response.text

# BeautifulSoup을 사용하여 HTML 파싱
soup = BeautifulSoup(html_content, "html.parser")

# 원하는 요소를 선택하여 가져오기
for item in soup.find_all("div", class_="common_sp_caption_tit line2"):
    print(item.text)
```
